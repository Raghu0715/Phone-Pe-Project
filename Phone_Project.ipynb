```python
import os
```

```python
path = "/content/Pulse/data/aggregated/transaction/country/india/state/"
agg_state_list = os.listdir(path)
#using a empty dictionary with key as a column and value as a empty list
clm = {'state':[],'Year':[],'quater':[],'Transaction_type':[],'Transaction_count':[], 'Transaction_amount':[]}
for i in agg_state_list:
  p_i = path+i+"/"
  agg_yr = os.listdir(p_i)
  for j in agg_yr:
    p_j = p_i + j + "/"
    agg_yr_list = os.listdir(p_j)
    for k in agg_yr_list:
      p_k = p_j + k  #i don't have any file inside this json file so i want to retrive data from this files and i am not adding "/"
      #open the each file
      Data = open(p_k,'r')
      D = json.load(Data)
      for  l in D['data']['transactionData']:
       Name = l['name']
       count = l['paymentInstruments'][0]['count']
       amount = l['paymentInstruments'][0]['amount']
       clm['Transaction_type'].append(Name)
       clm['Transaction_count'].append(count)
       clm['Transaction_amount'].append(amount)
       clm['state'].append(i)
       clm['Year'].append(j)
       clm['quater'].append(int(k.strip('.json')))
#creat a Data frame
pd.DataFrame(clm)
Agg_trans_india = pd.DataFrame(clm)
```

```python
Agg_trans_india.to_csv("aggregate_transaction_overall_india.csv")
```
